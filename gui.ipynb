```python
from fastai.vision.widgets import *
from fastai.vision.core import *
from fastai.learner import *
```

```python
import pathlib
temp = pathlib.WindowsPath
pathlib.WindowsPath = pathlib.PosixPath
```

```python
learner = load_learner('export.pkl')
```

```python
btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
prediction_txt = widgets.Label()
btn_run = widgets.Button(description='Predict')

def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    
    label, label_id, probs = learner.predict(img) # Do Inference
    
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(244,244))
    
    prediction_txt.value = f'Label: {label.capitalize()}, Prob: {probs[label_id]:.2%}'

btn_run.on_click(on_click_classify)


VBox([widgets.Label('Select your image!'), 
      btn_upload, btn_run, out_pl, prediction_txt])
```
